```python
from pandas.tools.plotting import parallel_coordinates
from sklearn.datasets import load_iris
import pandas as pd
%matplotlib inline
```

```python
def mapTarget(target):
    if target==0:
        return 'setosa'
    if target ==1:
        return 'versicolor'
    if target ==2:
        return 'virgincia'
    
```

```python
ir = load_iris()
ir.keys()
df = pd.DataFrame(ir['data'], columns=ir['feature_names'])
df['target'] = ir['target']
df['target_name'] = list(map(mapTarget, df['target']))
del df['target']
```

```python
df['target_name'].value_counts()
```

```python
parallel_coordinates(df[100:150], 'target_name')
```

```python
parallel_coordinates(df, 'target_name')
```
